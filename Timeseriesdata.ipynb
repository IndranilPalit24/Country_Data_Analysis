```python
import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn-v0_8')
```

```python
#importing the CSV file from the File Explorer to Python
data = pd.read_csv('E:\Kaggle DB\Extracted\ETH Data\ETH_day.csv')

#Changing the date in the data from the pd.to_datetime so that the datetime is not in string anymore and
#then sorting the Date values and changing the values in the actual dataset
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
#setting two variables namely price_date, to Date and price_close to Close values in the dataset
price_date = data['Date']

price_close = data['Close']
```

```python
#gcf = Date ticklabels often overlap, so it is useful to rotate them and right align them. 
#Also, a common use case is a number of subplots with shared x-axis where the x-axis is date data. 
#The ticklabels are often long, and it helps to rotate them on the bottom subplot and turn them off on other subplots, 
#as well as turn off xlabels. 


# automfmt - Date ticklabels often overlap, so it is useful to rotate them and right align them. 
#Also, a common use case is a number of subplots with shared x-axis where the x-axis is date data. 
#The ticklabels are often long, and it helps to rotate them on the bottom subplot and turn them off 
#on other subplots, as well as turn off xlabels.
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%m %d %Y')

plt.gca().xaxis.set_major_formatter(date_format)

plt.gcf().autofmt_xdate()


plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.plot(price_date, price_close)
plt.show()
```
