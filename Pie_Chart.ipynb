```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
```

```python
data = pd.read_csv('E:\Kaggle DB\Extracted\StackOverflowSurvey-2020\survey_results_public.csv')


plt.style.use('fivethirtyeight')

explode = [0,0,0.1,0,0]

language_counter = Counter()

for response in data['LanguageDesireNextYear'].dropna():
    language_counter.update(response.split(';'))

Languages = []
Popularity = []

for item in language_counter.most_common(5):
    Languages.append(item[0])
    Popularity.append(item[1])


plt.pie(Popularity, labels=Languages,explode = explode, shadow = True, autopct = '%1.1f%%' ,wedgeprops={'edgecolor': 'black'})

plt.tight_layout()

plt.show()

```
