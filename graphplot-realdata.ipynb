```python
#Path - 02
with open('E:\Kaggle DB\Extracted\StackOverflowSurvey-2020\survey_results_public.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()

    for row in csv_reader:
        language_counter.update(row['LanguageDesireNextYear'].split(';'))

print(language_counter)
```

```python
#We dont want all of the data to print the plot so we need only 15 languages from the table
#Now in the above data all the Languages and its numbers are jumbled up. If we want to seperate both of them in 
# a seperate list we need to declare them with a loop. Here it how it goes

Languages = []
Popularity = []

for item in language_counter.most_common(15):
    Languages.append(item[0])
    Popularity.append(item[1])

#We did a reverse here because we want to print the most used languages above. If we do not print the reverse 
#then the most used  programming langauge will be shown in the bottom
Languages.reverse()
Popularity.reverse()


    #Now that we have seperated the Languages in one list and Popularity in another list we need to plot them
    #barh meaning the bar chart changed from vertical to horizontal
plt.barh(Languages, Popularity)


#Plot Lengend
plt.title("Most Desired Languages")
plt.xlabel("Programming languages")
#plt.ylabel("Desire")

```

```python
#Now we need to do the same thing but with pandas
data = pd.read_csv('E:\Kaggle DB\Extracted\StackOverflowSurvey-2020\survey_results_public.csv', index_col = ['Respondent'])


language_counter = Counter()

#dropna - Because there are always some NaN valuees in the list. I tried this with ChatGPT. As of 05-02-2023 chatGPT is up and
#running
for response in data['LanguageDesireNextYear'].dropna():
    language_counter.update(response.split(';'))

Languages = []
Popularity = []

for item in language_counter.most_common(15):
    Languages.append(item[0])
    Popularity.append(item[1])


Languages.reverse()
Popularity.reverse()

plt.barh(Languages, Popularity)

plt.title("Most Desired Languages")
plt.xlabel("Programming languages")





```
