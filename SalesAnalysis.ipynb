#### Task #01 Merging 12 months of seperate data


```python
#creating a list comprehension for file and listing all the CSV files in the Sales Data Folder
files = [file for file in os.listdir('E:\Kaggle DB\Extracted\Pandas-Data-Science-Tasks-master\SalesAnalysis/Sales_Data')]


all_Months_data = pd.DataFrame()    #creating an empty data frame

#now creating a for loop with file as iterator in files and then reading all the csv_files one by one from 2019(April) to 2019(September)
for file in files:
    df = pd.read_csv("E:\Kaggle DB\Extracted\Pandas-Data-Science-Tasks-master\SalesAnalysis/Sales_Data/"+file)
    
    #Now concatenating the all_Months_data in new DataFrame to the df(dataframe)
    all_Months_data = pd.concat([all_Months_data,df])


## We list the files from os.listdir and listing all the CSV files then creating a new Empty DataFrame and 
# then having a for loop to read all the files from 2019 April to 2019 September
# and then concatenating all the data to the new DataFrame and then changing it to CSV and saving it to all_data.csv folder


all_Months_data.to_csv("all_data.csv", index=False)
```

#### Read in the updated DataFrame

```python
#reading the CSV data of all_data instead of copy pasting the entire upper row

all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Clean up the Data!

#### Drop Rows on NaN

```python
# We found some NAN here and wanted to remove all the rows containing NAN as we dont have a use of any of that so we use a
# variable called as nan_df and we select our all_data inside of the list we selected if there is NAN rows in all_data sets
# with any and axis=1(axis=1 represents column wise search, if axis =0 we would have search row wise)
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df

#once we found all the rows that contain NAN in it we need to drop then with the below command as we used the csv file
# and dropna and how is set to all that means it would drop all rows that has NAN, if we used any instead of all we would
# have dropped any row containing NAN
all_data = all_data.dropna(how='all')
all_data
```

##### Find OR and Delete it

```python
# We had some values in the Order Date which has months set as OR. Part of why Data is being cleaned everytime by a Data 
#Analyst. So now we removed by accessing Order Date column in all_data table and setting it like if we see values which has 
# Order date set as Or in 0 to 2 places in string we will not use them

all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

##### Convert Columns to the Correct type

```python
#all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')#make int
#all_data['Price Each'] = all_data['Price Each'].astype('float')
#all_data

## We created a copy of the original data in a variable and stopped from chaining and then we changed all the numeric 
#values from default string to numeric by accessing the table in which they were and changing then to to_numeric 
# and then checking by using all_data.head()

all_data_copy = all_data.copy()
all_data_copy['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #make Int
all_data_copy['Price Each'] = pd.to_numeric(all_data['Price Each']) #Make Float
all_data.head()
```

#### Augument Data with Additional Columns


##### Task 02: Add Month Column

```python
#all_data['Month'] = all_data['Order Date'].str[0:2]
#all_data['Month'] = all_data['Month'].astype('int32')
#all_data.head()

#all_data.loc[:, 'Month'] = all_data['Order Date'].str[0:2]
#all_data.loc[:, 'Month'] = all_data['Month'].astype('int32')


## Here we have the same copy of the original data and the changes made in the copy is affected in the original data too
# here we tried to change all the values in the month column from string to integer(numeric)

all_data_copy = all_data.copy()
all_data_copy['Month'] = all_data_copy['Order Date'].str[0:2]
all_data_copy['Month'] = pd.to_numeric(all_data_copy['Month'])
all_data.head()
```

#### Add Sales Column

```python
#all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
#all_data.head()

## We wanted to find the the maximum sales for a particular month so what we did that we used the copy of the variable 
# and then used it to multiply the values of the Quantity Ordered to Price Each.
all_data_copy = all_data.copy()
all_data_copy['Sales'] = all_data_copy['Quantity Ordered'] * all_data_copy['Price Each']
all_data.head()


# Once Done we sorted all values of Sales in ascending order
all_data_copy.sort_values('Sales', ascending =True, inplace=True)
all_data
```
