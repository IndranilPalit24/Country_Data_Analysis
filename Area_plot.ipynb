```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Extracting Data
data  =pd.read_csv('E:\Kaggle DB\Extracted\data.csv')

#Assigning data to a couple of variables
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
#Plotting a basic ine chart
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')

plt.plot(ages, py_salaries, label = 'Python')

#wiriting the legend so that all the labels in the plot will be displayed
plt.legend()

plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.show()
```

```python
#Now here lets start filling the areas below the plot. For that we need to use fill between after plotting the plot
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')

plt.plot(ages, py_salaries, label = 'Python')


plt.fill_between(ages, py_salaries, alpha=0.25) #This alpha=0.25 shows the area colour depth. Here we showed a light blue
#colour such that we can see the dotted lines

#wiriting the legend so that all the labels in the plot will be displayed
plt.legend()

plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.show()
```

```python
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')

plt.plot(ages, py_salaries, label = 'Python')

overall_median = 57287 #Median wherein the area will reverse

plt.fill_between(ages, py_salaries,overall_median, alpha=0.25) #This alpha=0.25 shows the area colour depth. Here we showed a light blue
#colour such that we can see the dotted lines

#wiriting the legend so that all the labels in the plot will be displayed
plt.legend()

plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.show()
```

```python
plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')

plt.plot(ages, py_salaries, label = 'Python')

overall_median = 57287 #Median wherein the area will reverse

plt.fill_between(ages, py_salaries,overall_median,where=(py_salaries> overall_median),interpolate=True, alpha=0.25) #This alpha=0.25 shows the area colour depth. Here we showed a light blue
#colour such that we can see the dotted lines. interpolate and where is to change the colours for the area plot above
#and below

plt.fill_between(ages, py_salaries,overall_median,where=(py_salaries<= overall_median),interpolate=True, color='red', alpha=0.25)

#wiriting the legend so that all the labels in the plot will be displayed
plt.legend()

plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.show()
```
